```python
from sklearn.model_selection import train_test_split

from sklearn.linear_model import LogisticRegressionCV
from sklearn.neighbors import KNeighborsClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.svm import SVC

from sklearn.decomposition import PCA
from sklearn.cluster import KMeans

from sklearn.model_selection import GridSearchCV

from sklearn.metrics import accuracy_score, confusion_matrix, classification_report
```

```python
X = df.drop("prognosis", axis=1)
y = df["prognosis"]
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.33, random_state=42)
```

```python
y_train.shape
```

```python
lr = LogisticRegressionCV()
lr.fit(X_train, y_train)
predictions = lr.predict(X_test)
```

```python
print(accuracy_score(y_test, predictions))
```

```python
#print(X.columns)
user_input = [


"vomiting",
"high_fever",
"sweating",
"headache",
"nausea",
"diarrhoea"

]

new_query = []
for i in X.columns:
    if i in user_input:
        new_query.append(1)
    elif i not in user_input:
        new_query.append(0)

new_array = np.array(new_query)
#print(new_array.shape)
#print(X_test.loc[12].shape)
doctor = lr.predict(new_array.reshape(1,-1))
probability = lr.predict_proba(new_array.reshape(1,-1))
print(doctor)
#print(probability)
#print(lr.classes_)
all_probabilities = {}
for index, proba in enumerate(probability.flatten()):
    all_probabilities[float(proba*100)] = lr.classes_.flatten()[index]
#print(np.sort(probability)[-1][::-1][:5])

for i in sorted(all_probabilities.items(), reverse=True)[:5]:
    print(i)
```
