```python
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity

def cosine_similarity_with_list(target_string, list_of_strings):
    vectorizer = TfidfVectorizer()
    tfidf_matrix = vectorizer.fit_transform([target_string] + list_of_strings)
    target_vector = tfidf_matrix[0]
    similarities = cosine_similarity(target_vector, tfidf_matrix[1:]).flatten()
    maximum = 0
    index = 0
    for i,j in enumerate(similarities.tolist()):
        if j > maximum:
            maximum = j
            index = i
    print(target_string, list_of_strings[index], maximum)
    return list_of_strings[index]
```

```python
def ml_dummies_create(counts, num_diseases, name_disease):

    matrix = np.zeros((counts, num_diseases))
    print(f"New matrix shape (number of cases, different symptoms): {matrix.shape}")
    print(f"Name of disease: {name_disease}")

    ones_count = [int(i) for i in new_frame3[new_frame3["diseases"]
    ==name_disease]["new_counts"].tolist()]
    print(f"Count of symptoms in cases: {ones_count}")

    for i, count in enumerate(ones_count):
        if count > counts:
            raise ValueError(f"Cannot place more ones ({count}) than the number of columns in row {i}.")
        indices = np.random.choice(counts, count, replace=False)
        matrix[indices, i] = 1

    df = pd.DataFrame(matrix, columns=new_frame3[new_frame3["diseases"]==name_disease]["symptoms"].tolist())
    df["prognosis"] = name_disease

    return df
```
